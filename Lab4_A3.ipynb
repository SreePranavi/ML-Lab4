```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate 20 random data points for x and y, values between 1 and 10
x = np.random.uniform(1, 10, 20)
y = np.random.uniform(1, 10, 20)

# Randomly assign each point to either class 0 or class 1
classes = np.random.choice([0, 1], size=20)

# Create a scatter plot
plt.figure(figsize=(8, 6))

# Plot class 0 points in blue
plt.scatter(x[classes == 0], y[classes == 0], color='blue', label='Class 0')

# Plot class 1 points in red
plt.scatter(x[classes == 1], y[classes == 1], color='red', label='Class 1')

# Add labels and title
plt.xlabel('Feature x')
plt.ylabel('Feature y')
plt.title('Scatter Plot of Training Data')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
