```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

def generate_training_data():
    np.random.seed(42)
    x_train = np.random.uniform(1, 10, 20)
    y_train = np.random.uniform(1, 10, 20)
    classes_train = np.random.choice([0, 1], size=20)
    return np.column_stack((x_train, y_train)), classes_train

def generate_test_data():
    x_test = np.arange(0, 10.1, 0.1)
    y_test = np.arange(0, 10.1, 0.1)
    return np.array(np.meshgrid(x_test, y_test)).T.reshape(-1, 2)

def classify_test_data(X_train, y_train, X_test, k=3):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    return knn.predict(X_test)

def plot_data(X_train, y_train, X_test, y_pred):
    plt.figure(figsize=(10, 8))
    plt.scatter(X_test[y_pred == 0][:, 0], X_test[y_pred == 0][:, 1], color='blue', label='Class 0', alpha=0.5)
    plt.scatter(X_test[y_pred == 1][:, 0], X_test[y_pred == 1][:, 1], color='red', label='Class 1', alpha=0.5)
    plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', edgecolor='black', label='Training Class 0', s=100)
    plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', edgecolor='black', label='Training Class 1', s=100)
    plt.xlabel('Feature x')
    plt.ylabel('Feature y')
    plt.title('KNN Classification (k=3) of Test Data')
    plt.legend()
    plt.show()

def main():
    X_train, y_train = generate_training_data()
    X_test = generate_test_data()
    y_pred = classify_test_data(X_train, y_train, X_test, k=3)
    plot_data(X_train, y_train, X_test, y_pred)
    print("Plot generated showing the KNN classification results with k=3.")

if __name__ == "__main__":
    main()
```
